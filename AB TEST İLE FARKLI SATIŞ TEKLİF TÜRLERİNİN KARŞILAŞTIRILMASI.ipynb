## A/B TEST İLE FARKLI SATIŞ TEKLİF TÜRLERİNİN KARŞILAŞTIRILMASI


### İŞ PROBLEMİ

Yakın zamanda bir sosyal medya platformu, mevcut teklif verme türüne (maksimum teklif verme) alternatif olarak yeni bir teklif verme türünü (ortalama teklif verme) kullanıma sundu.
Bu platformun müşterilerinden biri, bu yeni türü test etmeye ve mevcut olanla karşılaştırmaya karar verdi. 
İstemci, izleyicilerini rastgele olarak test ve kontrol grupları olarak eşit büyüklükte iki gruba ayırdı.
Veriler iki grubun birer aylık gözlem sonuçlarından oluşmaktadır. Bu çalışmada, müşterim seçilen grupları karşılaştırmamı istediği için A/B Test yöntemini kullanarak verileri analiz edeceğim. 

#### A/B için hipotezler

HO: Ortalama teklif verme ile maksimum teklif verme arasında satın alma sayısı açısından istatistiksel olarak anlamlı bir fark yoktur.

H1: Ortalama teklif ile maksimum teklif arasında satın alma sayısı açısından istatistiksel olarak anlamlı bir fark vardır.

#### İstatistiksel analiz yöntemleriyle normallik varsayımı testi 

```python
#Shapiro testi için kütüphane
from scipy import stats
```

##### Normallik Testi için Hipotezler
HO: Örnek normal dağılmıştır

H1: Örnek normal dağılmamış

```python
### Shapiro testi kullanılarak normallik testi
from scipy.stats import shapiro
```

```python
shapiro(A_B.df_A_nop)
```

- p değeri (0,97) > alfa (0,05) olduğundan H0'ı reddedemeyiz.
- Bu sonuç normallik varsayımının karşılandığını göstermektedir.

```python
shapiro(A_B.df_B_nop)
```

- p değeri (0.02) < alpha (0.05) olduğundan H0'ı reddediyoruz.
- Bu sonuç normallik varsayımının karşılanmadığını göstermektedir.
- Bu sonuç nedeniyle BAĞIMSIZ İKİ ÖRNEKLEM TESTİ KULLANILAMAMAKTADIR.

#### Varyans Testinin Homojenliği için Hipotezler
HO: Grupların varyansları eşittir (veya varyans gruplar arasında eşittir)

H1: Grupların varyansları eşit değil (veya varyans gruplar arasında eşit değil)

```python
#Levene Testi ile varyans homojenliği testi
stats.levene(A_B.df_A_nop, A_B.df_B_nop)
```

- p değeri (0,23) > alfa (0,05) olduğundan H0'ı reddedemeyiz.
- Bu sonuç varyans homojenliğinin sağlandığını göstermektedir.

#### Normal dağılım varsayımı karşılanamadığı için bağımsız iki örneklem T Testini kullanamıyoruz. Burada A/B Testi olarak  Mann Whitney U Testi kullanacağız

```python
#Mann Whitney U Testi
stats.mannwhitneyu(A_B.df_A_nop, A_B.df_B_nop)
```

```python
stats.mannwhitneyu(A_B.df_A_nop, A_B.df_B_nop)
```

- Bu sonuç, p değeri (0,19) > alfa değeri (0,05) olduğundan gruplar arasında istatistiksel olarak anlamlı bir fark olmadığını göstermektedir. Bu nedenle H0'ı reddedemeyiz.

 ## SONUÇ
    
Grupları karşılaştırmak için A/B Testi olarak Bağımsız İki Örnek T Testini kullanmak istedim; ancak, varyans homojenliği sağlanmış olsa da, Ortalama Teklif grubu (B grubu) için normallik varsayımı karşılanmamıştır.
Bu nedenle grupları karşılaştırmak için parametrik olmayan Bağımsız İki Örneklem Testi olarak Mann Whitney U Testi yapılmıştır. Yani grupları karşılaştırmak için başka bir A/B Testi kullandım.

A/B Testi (Mann Whitney U Testi) sonucu, p değeri (0,19) > alfa değeri (0,05) nedeniyle gruplar arasında istatistiksel olarak anlamlı fark olmadığını gösterdi.Bu nedenle H0'ı reddedemedik.
Diğer bir deyişle, iki grup için satın alma sayılarının dağılım şekilleri benzer olmadığından, satın alma sayısı açısından maksimum teklif verme (ortalama = 507) ve ortalama teklif verme (ortalama = 486) arasında istatistiksel olarak anlamlı bir farklılık olmadığını söyleyebiliriz.
(ya da maksimum teklif verme ile ortalama teklif verme arasında satın alma sayısı açısından istatistiksel olarak anlamlı bir fark olmadığını söyleyebiliriz). Yani, yeni teklif verme türü istatistiksel olarak anlamlı bir fark yaratmaz.
Mevcut A/B Testi sonucuna göre  müşterimize,şu anda yeni teklif türü kullanmasına gerek olmadığını tavsiye edebiliriz. 

Analizin sınırlaması: Her nekadar A/B Testi sonucu iki teklif şekli arasında istatistiksel olarak anlamlı bir fark olmadığını gösterse de, teklif verme şekillerinin karşılaştırıldığı yukarıdaki şekli de incelememizde fayda var. Bu grafik gözlemin 20. gününden itibaren ortalama teklif türünde (Test Kampanyası) satın alma sayısı artmaya başladığından dolayı, daha kesin sonuçlar elde edebilmek için 1 aydan daha uzun süreleri kapsayan kıyaslamaların gerekli olduğunu göstermektedir.
